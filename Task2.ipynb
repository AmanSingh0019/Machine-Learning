```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('/content/IMDb Movies India.csv', encoding='latin-1')
```

```python
data.info()
```

```python
data.head()
```

```python
data.describe()
```

```python
# Define features and target
features = ['Year', 'Duration', 'Votes']
target = 'Rating'
```

```python
# Convert 'Year', 'Duration', and 'Votes' to numerical, handling errors
data['Year'] = pd.to_numeric(data['Year'].astype(str).str.strip('()'), errors='coerce')
data['Duration'] = pd.to_numeric(data['Duration'].astype(str).str.extract('(\d+)', expand=False), errors='coerce')
data['Votes'] = pd.to_numeric(data['Votes'].astype(str).str.replace(',', ''), errors='coerce')

```

```python
# Handle missing values
data = data.dropna(subset=features + [target])
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    data[features], data[target], test_size=0.20, random_state=2549
)
X_test
```

```python
# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
print(y_pred)
```

```python
# Evaluate the model
import numpy as np
mse = mean_squared_error(y_test, y_pred)
# Calculate R-squared
r2 = 1 - (mse / np.var(y_test))

# Calculate RMSE
rmse = np.sqrt(mse)

# Calculate NRMSE (assuming 'Rating' is your target variable)
target_range = data['Rating'].max() - data['Rating'].min()
nrmse = (rmse / target_range) * 100

# Print the metrics
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Normalized RMSE (NRMSE): {nrmse:.2f}%")
```
